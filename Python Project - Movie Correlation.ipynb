```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create


# Read in the data

df = pd.read_csv(r'C:\Users\otboa\Downloads\movies.csv')
```

```python
# Investigate for missing values
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Change data type of "budget" and "gross" columns

df['budget'] = df['budget'].replace([np.inf, -np.inf], np.nan)  # Replace inf with NaN
df['budget'] = df['budget'].fillna(0)  # Replace NaN with 0
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].replace([np.inf, -np.inf], np.nan)  # Replace inf with NaN
df['gross'] = df['gross'].fillna(0)  # Replace NaN with 0
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# Function to extract the correct release year from the "released" column
from datetime import datetime


# Function to check if a string contains a date
def is_date_string(s):
    try:
        # Try parsing the string with common date formats
        datetime.strptime(s.split('(')[0].strip(), '%B %d, %Y')
        return True
    except ValueError:
        try:
            datetime.strptime(s.split('(')[0].strip(), '%B %Y')
            return True
        except ValueError:
            return False

# Function to extract the year
def extract_year(date_str):
    if isinstance(date_str, str) and is_date_string(date_str):
        # Remove the country name in parentheses
        cleaned_date_str = date_str.split('(')[0].strip()
        
        # Try different date formats
        try:
            # Try the full format with month, day, and year
            date_obj = datetime.strptime(cleaned_date_str, '%B %d, %Y')
        except ValueError:
            try:
                # If day is missing, try the month and year format
                date_obj = datetime.strptime(cleaned_date_str, '%B %Y')
            except ValueError:
                return None  # Return None if the format doesn't match any expected pattern
        
        # Extract and return the year
        return date_obj.year
    else:
        return None  # Return None if the value is not a valid date string
        
# Apply the function to the 'released' column
df['yearcorrect'] = df['released'].apply(extract_year)

df
```

```python
# Sorting the data by "gross" column
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any duplicates

df.drop_duplicates()
```

## We want to check the correlation between Gross revenue and other columns 

```python
# Scatter plot budget vs gross


plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show()


```

# Correlation - Budget vs Gross Earnings

```python
# Drop non-numeric columns
numeric_df = df.select_dtypes(include='number')

# Calculate the correlation
correlation_matrix = numeric_df.corr()

# Display the correlation matrix
print(correlation_matrix)
```

```python
# Visualize the correlation

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Looking at Company vs Gross
# Convert company to numeric so we can run a correlation on it

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized
```

```python
# Calculate the correlation
correlation_matrix1 = df_numerized.corr()

sns.heatmap(correlation_matrix1, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```
